```python
import requests
import pandas as pd
import numpy as np

def get_options_data(access_token, instrument_key="NSE_INDEX|Nifty 50", expiry_date="2025-05-22"):
    url = 'https://api.upstox.com/v2/option/chain'
    params = {
        'instrument_key': instrument_key,
        'expiry_date': expiry_date
    }
    headers = {
        'Accept': 'application/json',
        'Authorization': f'Bearer {access_token}'
    }

    response = requests.get(url, params=params, headers=headers)
    response_json = response.json()
    
    if 'data' not in response_json or not response_json['data']:
        return pd.DataFrame(), "No data available"
    
    # Extract all data first
    data = response_json['data']
    
    # Create empty lists to store organized data
    rows = []
    
    for item in data:
        strike_price = item.get('strike_price')
        
        # Get CE data
        ce_data = item.get('ce', {})
        pe_data = item.get('pe', {})
        
        # Create a row for this strike price
        row = {
            'strike_price': strike_price,
            'ce_ltp': ce_data.get('last_price', np.nan),
            'ce_iv': ce_data.get('implied_volatility', np.nan),
            'ce_delta': ce_data.get('delta', np.nan),
            'ce_gamma': ce_data.get('gamma', np.nan),
            'ce_theta': ce_data.get('theta', np.nan),
            'ce_vega': ce_data.get('vega', np.nan),
            'ce_oi': ce_data.get('open_interest', np.nan),
            'ce_volume': ce_data.get('volume', np.nan),
            'pe_ltp': pe_data.get('last_price', np.nan),
            'pe_iv': pe_data.get('implied_volatility', np.nan),
            'pe_delta': pe_data.get('delta', np.nan),
            'pe_gamma': pe_data.get('gamma', np.nan),
            'pe_theta': pe_data.get('theta', np.nan),
            'pe_vega': pe_data.get('vega', np.nan),
            'pe_oi': pe_data.get('open_interest', np.nan),
            'pe_volume': pe_data.get('volume', np.nan)
        }
        rows.append(row)
    
    # Create DataFrame and sort by strike price
    df = pd.DataFrame(rows)
    df = df.sort_values('strike_price')
    
    return df, "Data extracted successfully"

# Function to filter options data for specific strike prices
def filter_strike_prices(df, min_strike=None, max_strike=None, specific_strikes=None):
    if df.empty:
        return df
    
    if specific_strikes:
        return df[df['strike_price'].isin(specific_strikes)]
    
    filtered_df = df.copy()
    if min_strike is not None:
        filtered_df = filtered_df[filtered_df['strike_price'] >= min_strike]
    if max_strike is not None:
        filtered_df = filtered_df[filtered_df['strike_price'] <= max_strike]
    
    return filtered_df

# Example usage
access_token = "eyJ0eXAiOiJKV1QiLCJrZXlfaWQiOiJza192MS4wIiwiYWxnIjoiSFMyNTYifQ.eyJzdWIiOiI2WkJGV00iLCJqdGkiOiI2ODJhMjEzOWUxM2E3ODVjZTkzNDRlNWUiLCJpc011bHRpQ2xpZW50IjpmYWxzZSwiaXNQbHVzUGxhbiI6ZmFsc2UsImlhdCI6MTc0NzU5MTQ4MSwiaXNzIjoidWRhcGktZ2F0ZXdheS1zZXJ2aWNlIiwiZXhwIjoxNzQ3NjA1NjAwfQ.A11DtwS7wn76XY4jzI8TiWfHc-2YCbiVXffxeLHUxV4"

# Get all data
options_df, status = get_options_data(access_token)

# Example 1: Get specific strikes (22000, 22500, 23000)
specific_strikes = [25000, 24950]
specific_df = filter_strike_prices(options_df, specific_strikes=specific_strikes)

# Example 2: Get a range of strikes
range_df = filter_strike_prices(options_df, min_strike=24000, max_strike=25200)

# Display formatting for better readability
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("Status:", status)
print("\nOptions data for specific strikes:")
display(specific_df)

# Alternative: Find ATM options (closest to current spot price)
# You would need to get the current price from another API call or source
# For demonstration, let's assume it's around 22500
atm_strike = 22500
atm_df = options_df.iloc[(options_df['strike_price'] - atm_strike).abs().argsort()[:5]]
print("\nNear ATM options:")
display(atm_df)
```
